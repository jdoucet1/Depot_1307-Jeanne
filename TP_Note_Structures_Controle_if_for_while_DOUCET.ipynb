## Exercice 1 (3 pts)

Écrire une fonction qui reçoit un nombre `x` et retourne :
- `'Positif'` si `x > 0`
- `'Négatif'` si `x < 0`
- `'Zéro'` sinon



```python
# Définition de la fonction
def positif_ou_negatif(x):
    if x > 0:
        print("Positif")
    elif x < 0:
        print("Négatif")
    else:
        print("Zéro")

# Appel de la fonction
x = input("Veuillez entrer un nombre :")
x = float(x)
positif_ou_negatif(x)
```

## Exercice 2 (3 pts)

Écrire une fonction qui reçoit un âge (entier) et retourne :
- `'Enfant'` si âge < 12
- `'Adolescent'` si 12 ≤ âge ≤ 17
- `'Adulte'` si âge ≥ 18



```python
# Définition de la fonction
def fonction_age(age):
    while age < 0 or age > 125:
        print("Entrée non valide : veuillez entrer une valeur entre 0 et 125.")
        age = input("Veuillez entrer votre âge :")
        age = int(float(age))

    if age < 12:
        print("Enfant")
    elif age <= 17:
        print("Adolescent")
    else:
        print("Adulte")

# Appel de la fonction
age = input("Veuillez entrer votre âge :")
age = int(float(age))
fonction_age(age)
```

## Exercice 3 (3 pts)

Écrire une fonction qui reçoit une note sur 100 et retourne :
- `'Échec'` si note < 60
- `'Passable'` si 60–69
- `'Bien'` si 70–79
- `'Très bien'` si ≥ 80



```python
# Définition de la fonction
def statut_note(note):
    while note < 0 or note > 100:
        print("Entrée non valide : veuillez entrer une valeur entre 0 et 100 inclusivement.")
        note = input("Veuillez entrer la note obtenue :")
        note = float(note)
    
    if note < 60:
        print("Échec")
    elif note < 70:
        print("Passable")
    elif note < 80:
        print("Bien")
    else:
        print("Très bien")

# Appel de la fonction
note = input("Veuillez entrer la note obtenue :")
note = float(note)
statut_note(note)
```

## Exercice 4 (2 pts)

Écrire une fonction qui reçoit deux nombres `a` et `b` et retourne :
- le plus grand des deux
- ou la chaîne `'Égaux'` s’ils sont égaux



```python
# Définition de la fonction
def comparaison_nombres(a,b):
    if a > b:
        print(f"Le plus grand des deux nombres est {a}.")
    elif b > a:
        print(f"Le plus grand des deux nombres est {b}.")
    else:
        print("Les deux nombres sont égaux.")

# Appel de la fonction
a = input("Veuillez entrer un premier nombre :")
a = float(a)
b = input("Veuillez entrer un second nombre :")
b = float(b)
comparaison_nombres(a,b)
```

## Exercice 5 (4 pts)

Écrire une fonction qui reçoit un entier `n` et retourne une **liste de deux chaînes** :
1) `'Pair'` ou `'Impair'`
2) `'Multiple de 3'` ou `'Pas multiple de 3'`



```python
# Définition de la fonction
def pair_impair_multiple3(n):
    if n % 2 == 0:
        print("Pair")
    else:
        print("Impair")
    
    if n % 3 == 0:
        print("Multiple de 3")
    else:
        print("Pas multiple de 3")

# Appel de la fonction
n = input("Veuillez entrer un nombre entier :")
n = int(float(n))
pair_impair_multiple3(n)
```

## Exercice 10 (2 pts)

Écrire une fonction qui reçoit une chaîne `mot` et retourne la liste de ses caractères (un caractère par élément) en utilisant une boucle `for`.


```python
# Définition de la fonction
def chaine_mot(chaine):
    long = len(chaine)
    for i in range(0,long,1):
        print(chaine[i])

# Appel de la fonction
chaine = input("Veuillez entrer un ou plusieurs mots à décomposer :")
chaine = str(chaine)
chaine_mot(chaine)
```

## Exercice 12 (2 pts)

Écrire une fonction qui reçoit une liste de nombres saisis successivement et retourne la somme jusqu’à rencontrer un 0.

Exemple : `[5, 2, -1, 0, 100]` -> somme = `6`.

Consigne : utiliser une boucle `while`.




```python
def somme_liste(liste):
    somme = 0
    i = 0 # agira comme compteur
    while i < len(liste): # On aurait aussi pu utiliser une boucle for à la place
        somme = somme + liste[i]
        i += 1
        if liste[i] == 0:
            break
    return somme

# Appel de la fonction
liste = [5,2,-1,0,100]
res = somme_liste(liste)
print(f"La somme de tous les nombres saisis avant de rencontrer un 0 est {res}.")
```
